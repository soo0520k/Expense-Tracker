```python
sg.change_look_and_feel('TealMono')

layout = [      
          [sg.Text('Please enter your Name, Email Address, Phone Number')],   
    
          [sg.Text('Name', size=(15, 1)), sg.InputText('name')],      
    
          [sg.Text('Email Address', size=(15, 1)), sg.InputText('address')], 
    
          [sg.Text('Phone Number', size=(15, 1)), sg.InputText('phone')],   
    
          [sg.Submit(), sg.Cancel()]      
         ]

user_info=[]

window = sg.Window('Smart Money').Layout(layout)     

button, values = window.Read()

print(button, values[0], values[1], values[2])

user_info.append(values[0])

user_info.append(values[1])

user_info.append(values[2])

user_info

filepath="User.xlsx"

wb = load_workbook(filepath)

sheet = wb.active

sheet.cell(row=1, column=1).value= 'Expense Amount'

sheet.cell(row=1, column=2).value= 'Expense Date'

sheet.cell(row=1, column=3).value= 'Expense Type'

sheet.cell(row=1, column=4).value= 'Expense Description'

wb.save(filepath)

window.Close()

layout = [            
    [sg.Text('Smart Money collects your expenses and analyzes your spendings')],   
    
    [sg.Text('Please provide your expense amount, date, expense type and descriptions')], 
    
    [sg.Text('Amount', size=(15, 1)),sg.InputText(float())],
    
    [sg.Text('Date', size=(15, 1)),sg.InputText('MM/DD/YYYY')],
    
    [sg.Text('Expense Type', size=(15, 1)),sg.Listbox(values=('Restaurants & Dining','Shopping & Entertainment','Travel','Transportation', 'Groceries', 'Health', 'Misc'), size=(30, 3))],
    
    [sg.Text('Expense Description', size=(15, 1)),sg.InputText()], 
    
    [sg.Submit(), sg.Cancel()]      
    ]

window = sg.Window('Smart Money').Layout(layout)

sg.Popup(button, values)

expense=[]

while True:
    
    button, values = window.read()
    
    if button in (None, 'Cancel'):
            
        break
        
    print(str(values))
    
    expense.append(values)

sheet=wb.active

row= sheet.max_row

for j in range(len(expense)):
    
    for i in range(4):
        
        sheet.cell(row=j+row+1, column=i+1).value= str(expense[j][i])

wb.save(filepath)

window.Close()
```

```python
user_file = pd.read_excel('User.xlsx')
```

```python
for i in range (len(user_file['Expense Type'])):  
    
    for char in "['']":
        
        user_file['Expense Type'][i] = user_file['Expense Type'][i].replace(char,'')
user_file
```

```python
#Summary of Expenses

sg.change_look_and_feel('TealMono')

summary= user_file.groupby(by=['Expense Type'])['Expense Amount'].sum()

layout = [[sg.Text('Summary of My Expenses', font='Any 15', key= "-bold-")],
          
          [sg.Text(summary.to_string(index=True), font='Any 10')], 
          
          [sg.OK()]]

window = sg.Window('Smart Money, Summary',layout, finalize=True)

event, values = window.read()

window.close()
```

```python
#Pie Chart of the Summary
sg.change_look_and_feel('TealMono')

matplotlib.use('TkAgg')

plt.pie(summary, labels = summary.index,startangle=90 ,shadow = False ,autopct='%1.1f%%')

plt.title('Summary of My Expenses',color='Black')

plt.axis('equal')

plt.rcParams['font.sans-serif'] = 'Calibri'

plt.tight_layout()

fig = plt.gcf()     

figure_x, figure_y, figure_w, figure_h = fig.bbox.bounds

def draw_figure(canvas, figure, loc=(0, 0)):

    figure_canvas_agg = FigureCanvasTkAgg(figure, canvas)

    figure_canvas_agg.draw()

    figure_canvas_agg.get_tk_widget().pack(side='top', fill='both', expand=1)

    return figure_canvas_agg

layout = [[sg.Text('Summary of My expenses', font='Any 18')],

          [sg.Canvas(size=(figure_w, figure_h), key='canvas')],

          [sg.OK()]]


window = sg.Window('Smart Money, Summary',

                   layout, finalize=True)

fig_canvas_agg = draw_figure(window['canvas'].TKCanvas, fig)

event, values = window.read()

window.close()
```
